```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random
```

```python
req = requests.get('https://editorial.rottentomatoes.com/guide/best-netflix-movies-to-watch-right-now/')
webpage = req.text
```

```python
type(webpage)
```

```python
with open("filename", "wb") as f:
    f.write(webpage.encode('utf-8'))
```

```python
soup = BeautifulSoup(webpage, 'html.parser')
```

```python
print(soup.prettify())
```

```python

movies_data = []

# Find all movie containers
movie_divs = soup.find_all('div', class_='row countdown-item')
for movie_div in movie_divs:
    # Extract movie rank
    rank = movie_div.find('div', class_='countdown-index-resposive').get_text(strip=True).replace("#", "")
    
    # Extract movie name and release year
    title_div = movie_div.find('div', class_='article_movie_title')
    name = title_div.find('a').get_text(strip=True)
    year = title_div.find('span', class_='start-year').get_text(strip=True).replace("(", "").replace(")", "")
    
    # Extract rating (Tomatometer score)
    rating = title_div.find('span', class_='tMeterScore').get_text(strip=True).replace("%", "")
        
    # Extract adjusted score
    adjusted_score = movie_div.find('div', class_='info countdown-adjusted-score').get_text(strip=True).replace("Adjusted Score:", "").strip().replace("%", "")
    
    # Extract critics consensus
    consensus_div = movie_div.find('div', class_='info critics-consensus')
    consensus = consensus_div.get_text(strip=True).replace("Critics Consensus:", "").strip() if consensus_div else None
    
    # Extract synopsis
    synopsis_div = movie_div.find('div', class_='info synopsis')
    synopsis = synopsis_div.get_text(strip=True).replace("Synopsis:", "").strip() if synopsis_div else None
    
    # Extract cast
    cast_div = movie_div.find('div', class_='info cast')
    cast = ", ".join([a.get_text(strip=True) for a in cast_div.find_all('a')]) if cast_div else None
    
    # Extract director
    director_div = movie_div.find('div', class_='info director')
    director = ", ".join([a.get_text(strip=True) for a in director_div.find_all('a')]) if director_div else None
    
    # Add data to the list
    movies_data.append({
        "Rank": rank,
        "Name": name,
        "Year": year,
        "Rating (%)": rating,
        "Adjusted Score (%)": adjusted_score,
        "Critics Consensus": consensus,
        "Synopsis": synopsis,
        "Cast": cast,
        "Director": director
        })

# Convert to DataFrame
df = pd.DataFrame(movies_data)

# Save to a CSV file
df.to_csv("best_movies_netflix.csv", index=False)
print("Data saved to 'best_movies_netflix.csv'")

```
