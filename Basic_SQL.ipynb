Q3(i).

SELECT COUNT(*) AS count

FROM Customers;

Q3(ii).

SELECT customer_id, COUNT(*) as count

FROM Customers

GROUP BY customer_id

HAVING COUNT(*) > 1;

Q3(iii).

SELECT country, COUNT(DISTINCT customer_id) AS numcustomers

FROM Customers

GROUP BY country

ORDER BY numcustomers DESC;

Q5(iii).

SELECT

    order_id,

    order_date,

    shipped_date,

    (shipped_date - order_date) AS gap

FROM Orders

WHERE shipped_date IS NOT NULL

AND (shipped_date - order_date) > 30

ORDER BY gap DESC;

Q6(i).

SELECT

    order_id,
    
    ROUND(SUM(unit_price * quantity)::NUMERIC,2) AS sub_total

FROM Order_Details

GROUP by order_id

ORDER by sub_total DESC

LIMIT 5;

Q6(ii).

    order_id,

    ROUND(SUM(unit_price * quantity)::NUMERIC,0) AS sub_total

FROM Order_Details

WHERE order_id IN(

    SELECT order_id

    FROM Order_Details

    GROUP BY order_id

    HAVING COUNT(DISTINCT product_id) >5
)

GROUP BY order_id

ORDER BY sub_total DESC;

Q9.

SELECT

    product_id,

    COUNT(DISTINCT order_id) AS num_order,

    ROUND(AVG(unit_price)::NUMERIC,2) AS avg_unit_price

FROM Order_Details

GROUP BY product_id

HAVING ROUND(AVG(unit_price)::NUMERIC,2) < 7

ORDER BY avg_unit_price ASC;
